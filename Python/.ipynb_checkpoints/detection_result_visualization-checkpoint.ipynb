```python
def plot_barh_by_df(df, x_col, y_col, ylabel, xlabel, fig_name=None, 
                    figsize=(16,12), fontsize=16, lr=0.1, tb=.0,label_stacked=False, cat_num=43, **kwargs):
    """Plots horizintal bar. It displays percentages and frequencies.
       As input it requires dataframe with columns for x and y"""

    fig, ax = plt.subplots(figsize=figsize)
    if len(y_col) == 2:
        df.plot(x=x_col, y=y_col, kind="barh", ax=ax, legend=False, stacked=True)
    else:
        df.plot(x=x_col, y=y_col, kind="barh", ax=ax, legend=False)
    ax.set_ylabel(ylabel, fontsize=fontsize)
    ax.set_xlabel(xlabel, fontsize=fontsize)
    ax.tick_params(axis='both', which='major', labelsize=fontsize)
    if label_stacked:
        labels_best = []
        labels_last = []
        number_of_images = []
        for count, i in enumerate(ax.patches):
            if count < cat_num:
                labels_best.append([i.get_width(), i.get_y()])
                number_of_images.append(df["total"].get(count)) #+df["test_set"].get(count)
            else:
                labels_last.append([i.get_width(), i.get_y()])
            
        i = 0    
        for best, last in zip(labels_best, labels_last):
                ax.text(best[0]+last[0]+lr, best[1]+tb, \
                        str(best[0].round(4)) + " | " +  str(last[0].round(4)) + " (" +str(number_of_images[i])+")",
                        fontsize=fontsize, **kwargs)
                i+=1
    else:
        for count, i in enumerate(ax.patches):
        # get_width pulls left or right; get_y pushes up or down
            ax.text(i.get_width()+lr, i.get_y()+tb, \
                    str(i.get_width()) + " ("+str(df["train_set"].get(count))+"|"+str(df["test_set"].get(count))+")",
                    fontsize=fontsize, **kwargs)


        
    plt.tight_layout()
    if fig_name:     
        plt.savefig("results/images/"+fig_name+".png", bbox_inches='tight')
```

```python
model_code = "model"
model_path = "results/Model.csv" 
model_df = pd.read_csv(model_path, sep=",")
model_df["mAP_score"]=np.where(model_df["mAP_score"]==-1,0,model_df["mAP_score"]).round(4)
print("Min mAP score: ", model_df["mAP_score"].min())
print("Max mAP score: ", model_df["mAP_score"].max())
# Reading train-test split for "Model1"
train_labels_model_dict = pd.read_csv("dataset/"+model_code+"/train_labels.csv")["class"].value_counts().to_dict()
test_labels_model_dict = pd.read_csv("dataset/"+model_code+"/test_labels.csv")["class"].value_counts().to_dict()
# Creating table for comparison of results and train-test split
model_df["train_set"] = model_df["traffic_sign"].map(train_labels_model_dict).fillna(0).astype(int)
model_df["test_set"] = model_df["traffic_sign"].map(test_labels_model_dict).fillna(0).astype(int)
model_df_sorted = model_df.sort_values(by="mAP_score", ascending=False).reset_index(drop=True)
model_df_sorted["total"] = model_df_sorted["train_set"] + model_df_sorted["test_set"]
```

```python
# Plotting
fig_name = "model4_last_results"
plot_barh_by_df(model_df_sorted.sort_values(by="last_mAP_score", ascending=False).reset_index(drop=True).round(4),
                x_col="traffic_sign", y_col=["last_mAP_score"],
                ylabel="Traffic sign category", xlabel="mAP score (number of train images|number of test images)",
                fig_name=fig_name, figsize=(12,4), fontsize=16, lr=.001, tb=.16, label_stacked=False)
```
